```python
import os
from bs4 import BeautifulSoup
```

```python
dir = os.listdir("../data/mask")
dir = sorted(dir)
length = len(dir)
train_length = int(length*0.8)
output_file = open('../data/train.txt', 'w')
for i in range(train_length):
    with open('../data/mask/'+dir[i], 'r') as f:
        data = f.read()
    data = data.split(" ")
    boxes = []
    tmp = []
    for i in range(len(data)):
        if "filename" in data[i]:
            file_name = (data[i].split(">")[1].split("<")[0])
        if "xmin" in data[i]:
            tmp.append(data[i].split(">")[1].split("<")[0])
        if "ymin" in data[i]:
            tmp.append(data[i].split(">")[1].split("<")[0])
        if "xmax" in data[i]:
            tmp.append(data[i].split(">")[1].split("<")[0])
        if "ymax" in data[i]:
            tmp.append(data[i].split(">")[1].split("<")[0])
            boxes.append(tmp)
            tmp=[]

    print(file_name)
    
    s = file_name+" "
    for i in range(len(boxes)):
        for j in range(len(boxes[i])):
            s+=boxes[i][j]+" "
        s+="0 "
    s+="\n"
    output_file.write(s)
#     break
output_file.close()


output_file = open('../data/valid.txt', 'w')
for i in range(train_length,length):
    with open('../data/mask/'+dir[i], 'r') as f:
        data = f.read()
    data = data.split(" ")
    boxes = []
    tmp = []
    for i in range(len(data)):
        if "filename" in data[i]:
            file_name = (data[i].split(">")[1].split("<")[0])
        if "xmin" in data[i]:
            tmp.append(data[i].split(">")[1].split("<")[0])
        if "ymin" in data[i]:
            tmp.append(data[i].split(">")[1].split("<")[0])
        if "xmax" in data[i]:
            tmp.append(data[i].split(">")[1].split("<")[0])
        if "ymax" in data[i]:
            tmp.append(data[i].split(">")[1].split("<")[0])
            boxes.append(tmp)
            tmp=[]

    print(file_name)
    
    s = file_name+" "
    for i in range(len(boxes)):
        for j in range(len(boxes[i])):
            s+=boxes[i][j]+" "
        s+="0 "
    s+="\n"
    output_file.write(s)
#     break
output_file.close()
```
